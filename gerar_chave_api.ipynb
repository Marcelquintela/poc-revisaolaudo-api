# Gerar e salvar API Key em .env
Este notebook gera uma chave segura e salva em `.env` na pasta do projeto.

```python
import secrets
```

```python
# Gera uma chave hex de 64 caracteres (32 bytes)
api_key = secrets.token_hex(8)
print('Chave gerada:', api_key)
```

# Salvar em .env

```python

content = f'API_KEY_REVISAO_LAUDO={api_key}\n'
with open('.env', 'w') as f:
	f.write(content)

print('.env criado/atualizado com sucesso')
```
